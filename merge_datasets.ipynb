```python
#Import Packages
import pandas as pd
import numpy as np
import math
```

```python
important_data = pd.read_csv("Gene_importance_complete.csv")
important_data
```

```python
expression_data = pd.read_csv("combinedFiltered.csv")
expression_data
```

```python
#Create Expression Level Dataframe
classes = ['Gene','Blood_Expression','Breast_Expression','Kidney_Expression','Lung_Expression']
expression_levels = pd.DataFrame(np.zeros((769,5)),columns = classes)
expression_levels.iloc[:,0] = important_data.iloc[:,0]
expression_levels
```

```python
#Fill in Expression Level Dataframe
X = expression_data.iloc[:,:-1].copy()
y = expression_data.iloc[:,-1].copy()
cancer_types = ['Blood','Breast','Kidney','Lung']
for i, cancer in enumerate(cancer_types):
    X_partial = X.loc[y == cancer].reset_index(drop=True)
    avg_values = np.mean(np.array(X_partial), axis = 0)
    expression_levels.iloc[:,i+1] = avg_values
expression_levels
```

```python
df1 = important_data
df2 = expression_levels.drop(['Gene'],axis=1)
df_concat = pd.concat([df1, df2], axis=1)
df_concat
```
